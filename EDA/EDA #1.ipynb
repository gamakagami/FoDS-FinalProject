```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")

df = pd.read_csv("merged_cleaned.csv")
```

```python
print(df.shape)
```

```python
plt.figure(figsize=(12, 8)) 
histogram = df["Average_Engagement_Rate"].plot(kind="hist", bins = 40, title="Average Engagement Rate Histogram (All Data)", edgecolor="Black", range=(0,270))
histogram.set_xlabel("Average Engagement Rate")
plt.show()
```

```python
df_nano = df[df["Followers"] <= 10000]

plt.figure(figsize=(12, 8)) 
histogram = df_nano["Average_Engagement_Rate"].plot(kind="hist", bins = 40, title="Average Engagement Rate Histogram (Nano)", edgecolor="Black", range=(0,270))
histogram.set_xlabel("Average Engagement Rate")

plt.show()
```

```python
df_micro = df[(df["Followers"] >= 10000) & (df["Followers"] <= 100000)]

plt.figure(figsize=(12, 8)) 
histogram = df_micro["Average_Engagement_Rate"].plot(kind="hist", bins = 40, title="Average Engagement Rate Histogram (Micro)", edgecolor="Black", range=(0,270))
histogram.set_xlabel("Average Engagement Rate")
plt.show()
```

```python
df_macro = df[(df["Followers"] >= 100000) & (df["Followers"] <= 1000000)]

plt.figure(figsize=(12, 8)) 
histogram = df_macro["Average_Engagement_Rate"].plot(kind="hist", bins = 40, title="Average Engagement Rate Histogram (Macro)", edgecolor="Black", range=(0,270))
histogram.set_xlabel("Average Engagement Rate")
plt.show()
```

```python
df_mega = df[(df["Followers"] >= 1000000)]

plt.figure(figsize=(12, 8)) 
histogram = df_mega["Average_Engagement_Rate"].plot(kind="hist", bins = 40, title="Average Engagement Rate Histogram (Mega)", edgecolor="Black", range=(0,270))
histogram.set_xlabel("Average Engagement Rate")
plt.show()
```
