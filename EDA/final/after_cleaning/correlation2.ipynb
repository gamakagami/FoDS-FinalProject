```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
```

```python
df = pd.read_csv('merged_z5.csv', encoding='latin1')

df["Image_Ratio"] = df["Posts_(image)"] / df["Posting_frequency_(last_60_days)"]
df["Video_Ratio"] = df["Posts_(videos)"] / df["Posting_frequency_(last_60_days)"]
df["Carousel_Ratio"] = df["Posts_(carousel)"] / df["Posting_frequency_(last_60_days)"]

df.rename(columns={"Average_Likes" : "Av_Likes"}, inplace=True)
df.rename(columns={"Average_Comments" : "Av_Comments"}, inplace=True)
df.rename(columns={"Average_Engagement_Rate" : "Av_EngRate"}, inplace=True)
df.rename(columns={"Engagement_Rate_(60_Days)" : "Av_EngRate(60Days)"}, inplace=True)
df.rename(columns={"Posting_frequency_(last_60_days)" : "Post_Freq(60Days)"}, inplace=True)

plt.figure(figsize=(12, 8))

df = df[["Av_EngRate", "Av_EngRate(60Days)", "Image_Ratio", "Video_Ratio", "Carousel_Ratio"]].copy()

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
