```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")

df = pd.read_csv("Macro_Data.csv")
```

```python
df.shape
```

```python
df.rename(columns={"Engagement Rate (60 Days)" : "Average Engagement Rate (60 Days)"}, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
# Checking for duplicate usernames
df["Username"].value_counts()
```

```python
ax = df["Average Engagement Rate"].value_counts().head(10).plot(kind="bar", title="Average Engagement Rate")
ax.set_xlabel("Average Engagement Rate")
ax.set_ylabel("Frequency")
```

```python
histogram = df["Average Engagement Rate"].plot(kind="hist", bins = 20)
histogram.set_xlabel("Average Engagement Rate")

histogram_recent = df["Average Engagement Rate (60 Days)"].plot(kind="hist", bins = 20)
histogram_recent.set_xlabel("Average Engagement Rate (60 Days)")
```

```python
sde = df["Average Engagement Rate (60 Days)"].plot(kind="kde")
sde.set_xlabel("Average Engagement Rate (60 Days)")
```

```python
df.plot(kind="scatter", x="Followers", y="Average Engagement Rate")
plt.ticklabel_format(style="plain", axis="x")
```

```python
original_columns = df.columns.tolist()

df.rename(columns={
    "Average Engagement Rate" : "AER",
    "Average Engagement Rate (60 Days)" : "AER (60D)",
    "Average Hashtags / Post (60 Days)" : "AH (60D)"
}, inplace=True)

sns.pairplot(df, vars=["Followers", "AER", "AER (60D)", "AH (60D)"])
df.columns = original_columns
df.head()
```

```python
df_corr = df[["Followers", "Average Likes", "Average Comments", "Average Engagement Rate", "Average Engagement Rate (60 Days)", "Posting frequency (last 60 days)"]].corr()

sns.heatmap(df_corr, annot=True)
```
